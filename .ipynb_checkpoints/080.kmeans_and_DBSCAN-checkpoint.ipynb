# K-Means Clustering

### Random 하게 생성된 toy dataset 으로 K-Means clustering test

make_blobs : 
```
    Input :
         n_samples :  cluster 에 균등 분할될 total data point 숫자  
         centers : generate 할 centroid 숫자  
         cluster_std : cluster 의 standard deviation    

    Output : 
        X: 생성된 sample 들  
        y: 각 sample 의 label
```
KMeans :  
> init : initialization method -> k-means++ (smart choosing of centroids)  
> n_clusters : k 값  
> n_init : 반복횟수  

DBSCAN :

>eps : epsilon (radius)  
>min_sample : minimum samples within the radius

````python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
````

````python
np.random.seed(101)
````

````python
centroidLocation = [[3,2], [1,-1],[-1,2]]

X, _ = make_blobs(n_samples=1500, centers=centroidLocation)
````

````python
plt.scatter(X[:,0], X[:,1], marker='.')
````

````python
print(X.shape)
print(X[:10])
````

````python
nclusters = 3
````

````python
k_means = KMeans(n_clusters=nclusters)
````

````python
k_means.fit(X)
````

````python
k_means.labels_
````

````python
centers = k_means.cluster_centers_
centers
````

````python
from matplotlib.colors import ListedColormap

# 뚜렷한 색상의 사용자 정의 컬러맵 생성
# 여기서는 빨간색, 초록색, 파란색을 각각의 클래스에 대한 데이터 포인트 색상으로 사용
colors_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# 연한 색상의 사용자 정의 컬러맵 생성
# 여기서는 연한 빨간색, 연한 초록색, 연한 파란색을 각각의 클래스에 대한 배경 색상으로 사용
colors_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
````

````python
plt.figure(figsize=(8,6))

# 클러스터의 수만큼 반복하여 각 클러스터의 데이터 포인트와 중심을 플로팅
for i in range(nclusters):
    # i번째 클러스터에 속하는 데이터 포인트를 선택
    members = k_means.labels_ == i
    # 선택된 데이터 포인트를 플로팅
    plt.plot(X[members, 0], X[members, 1], '.', 
             color=colors_light(i), markersize=10, label='Cluster {}'.format(i))
    # i번째 클러스터의 중심을 플로팅
    plt.plot(centers[i, 0], centers[i, 1], 'o', 
             color=colors_bold(i), markeredgecolor='k', markersize=20)

# 그래프의 제목을 설정
plt.title("KMeans Clustering")
# 범례를 표시
plt.legend()

# 그래프를 표시
plt.show()
````

# DBSCAN

- outlier 가 있는 data points를 생성하고 DBSCAN 알고리즘이 outlier를 검출하는지 test

````python
from sklearn.cluster import DBSCAN

X1, _ = make_blobs(n_samples=500, centers=[[-3,-3]])
X2, _ = make_blobs(n_samples=500, centers=[[3,3]])
X3 = np.random.rand(500, 2) * 3 + 4
X4 = np.random.randn(10, 2) * 3  #outlier

X1.shape, X2.shape, X3.shape, X4.shape
````

````python
plt.figure(figsize=(8, 6))
plt.scatter(X1[:, 0], X1[:, 1], marker='.')
plt.scatter(X2[:, 0], X2[:, 1], marker='.')
plt.scatter(X3[:, 0], X3[:, 1], marker='.')
plt.scatter(X4[:, 0], X4[:, 1], marker='.')
````

````python
X = np.vstack([X1, X2, X3, X4])
X.shape
````

````python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], marker='.')
````

````python
epsilon = 0.3
minimumSamples = 7

db = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(X)

labels = db.labels_

print(labels.shape)
print(labels)
print(db.core_sample_indices_.shape)
print(db.core_sample_indices_)
````

````python
unique_labels = set(labels)       # -1 : outlier
unique_labels
````

````python
# 색상 스펙트럼을 설정합니다. 이는 클러스터의 개수에 따라 색상을 동적으로 생성합니다.
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

# 그래프의 크기를 설정합니다.
plt.figure(figsize=(8, 6))

# 유니크한 라벨과 색상에 대해 반복하면서 각 클러스터의 데이터 포인트를 산점도로 그립니다.
for k, col in zip(unique_labels, colors):
    members = (labels == k) # 라벨 k에 해당하는 멤버들을 선택
    # 해당 멤버들의 데이터 포인트를 산점도로 그립니다.
    plt.scatter(X[members, 0], X[members, 1], color=col, marker='o', s=10)  # s는 마커의 크기

# 그래프의 제목을 설정합니다.
plt.title('DBSCAN Clustering')
# 그래프를 화면에 표시합니다.
plt.show()
````
